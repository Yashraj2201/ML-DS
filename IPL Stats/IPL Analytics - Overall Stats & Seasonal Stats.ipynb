```python
import pandas as pd
```

## Overall Stats

```python
stats1 = pd.read_csv("2008_2021_updated_ball.csv")
```

```python
stats1.head(10)
```

```python
stats2 = pd.read_csv("2008_2021_data_matches.csv")
```

```python
stats2.head(10)
```

```python
stats = pd.merge(stats1, stats2, on = 'id')
stats.columns
```

### Filtering Data - Before 2022

```python
stats['date'] = pd.to_datetime(stats['date'])
```

```python
stats = stats.query(" date < '2022-01-01'")
```

```python
stats.sort_values(by = 'date', ascending = True)
```

```python
overall_stats = stats[['batsman' , 'batsman_runs', 'inning', 'season', 'date','id']]
```

```python
overall_stats.dtypes
```

```python
overall_stats
```

```python
overall_stats_final = overall_stats.groupby("batsman", as_index = False).agg({'batsman_runs' : sum, 
                                                                       'id' : 'nunique'})
```

```python
overall_stats_final = overall_stats_final.sort_values('batsman_runs', ascending = False)
overall_stats_final.rename(columns = {'id' : 'innings', 'batsman_runs' : 'runs'}, inplace = True)
overall_stats_final.head(5)
```

## Seasonal Stats 

```python
stats.columns
```

```python
seasonal_stats = stats[['id', 'inning', 'batsman', 'batsman_runs','date','season']]
```

```python
seasonal_stats.head(5)
```

```python
seasonal_stats_final = seasonal_stats.groupby(["batsman", "season"], as_index = False).agg({'batsman_runs' : sum, 
                                                                       'id' : 'nunique'})
```

```python
seasonal_stats_final
```

```python
seasonal_stats_final = seasonal_stats_final.sort_values(by = ['season','batsman_runs'], ascending = [True, False])
```
