```python
import pandas as pd
```

## Overall Stats

```python
stats1 = pd.read_csv("2008_2021_updated_ball.csv")
```

```python
stats1.head(15)
```

```python
stats2 = pd.read_csv("2008_2021_data_matches.csv")
```

```python
stats2.head(15)
```

```python
overall_stats = pd.merge(stats1, stats2, on = 'id')
```

```python
overall_stats = overall_stats[['batsman' , 'batsman_runs', 'inning', 'season', 'date']]
```

```python
overall_stats
```

```python
overall_stats.dtypes
```

```python
overall_stats['date'] = pd.to_datetime(overall_stats['date'])
```

```python
filtered_overall_stats = overall_stats.query(" date < '2022-01-01'")
```

```python
filtered_overall_stats.sort_values(by = 'date', ascending = True)
```

```python
overall_stats_final = filtered_overall_stats.groupby("batsman", as_index = False).agg({'batsman_runs' : sum, 
                                                                       'inning' : sum})
```

```python
overall_stats_final = overall_stats_final.sort_values('batsman_runs', ascending = False)
```
