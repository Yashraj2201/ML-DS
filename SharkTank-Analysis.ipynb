```python
sharkTankData_df['Accepted Offer'].fillna(0, inplace=True)
sharkTankData_df['Male Presenters'].fillna(0, inplace=True)
sharkTankData_df['Female Presenters'].fillna(0, inplace=True)
sharkTankData_df['Transgender Presenters'].fillna(0, inplace=True)
```

```python
sharkTankData_df['Accepted Offer'] = sharkTankData_df['Accepted Offer'].astype('int64')
sharkTankData_df['Male Presenters'] = sharkTankData_df['Male Presenters'].astype('int64')
sharkTankData_df['Female Presenters'] = sharkTankData_df['Female Presenters'].astype('int64')
sharkTankData_df['Transgender Presenters'] = sharkTankData_df['Transgender Presenters'].astype('int64')
```

```python
#basic info
print("Total Seasons: ", sharkTankData_df['Season Number'].max() )
print("Total Episodes: ", sharkTankData_df['Episode Number'].count())
print("Total Company Pitches: ", sharkTankData_df['Startup Name'].nunique())
print("Total Offers Received: ", sharkTankData_df['Received Offer'].sum())
print("Total Offers Accepted: ", sharkTankData_df['Accepted Offer'].sum())
#print("Time Period: ")
```
